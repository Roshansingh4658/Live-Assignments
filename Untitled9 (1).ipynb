```python
import numpy as np

# Method 1: Using np.full
array1 = np.full((3, 3), 7)
print("Array 1 using np.full:\n", array1)

# Method 2: Using np.ones and multiplying
array2 = np.ones((3, 3)) * 7
print("Array 2 using np.ones:\n", array2)

# Method 3: Using np.tile
array3 = np.tile(7, (3, 3))
print("Array 3 using np.tile:\n", array3)
```

```python
import numpy as np

# np.array always creates a copy of the data
a = [1, 2, 3]
b = np.array(a)
print("np.array creates a copy:", b is a)  # False

# np.asarray doesn't create a copy if the input is already an ndarray
c = np.asarray(a)
print("np.asarray creates a copy:", c is a)  # False

# np.asanyarray is similar to np.asarray but passes through subclasses
d = np.asanyarray(a)
print("np.asanyarray creates a copy:", d is a)  # False
```

```python
random_int_array = np.random.randint(1, 11, (5, 6))
print("Original Array:\n", random_int_array)

# a) Extract all even integers from the array
even_integers = random_int_array[random_int_array % 2 == 0]
print("Even integers:\n", even_integers)

# b) Extract all odd integers from the array
odd_integers = random_int_array[random_int_array % 2 != 0]
print("Odd integers:\n", odd_integers)
```

```python
array_3d = np.random.randint(1, 11, (3, 3, 3))
print("3D Array:\n", array_3d)

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3d, axis=2)
print("Indices of max values along depth:\n", max_indices)

# b) Perform element-wise multiplication of between both arrays (assuming another 3D array)
array_3d_2 = np.random.randint(1, 11, (3, 3, 3))
multiplied_array = array_3d * array_3d_2
print("Element-wise multiplication:\n", multiplied_array)
```

```python
import pandas as pd

# Assuming df is your DataFrame
df = pd.DataFrame({'Phone': ['+1-800-555-5555', '800-555-5555', '(800)555-5555']})

# Removing non-numeric characters and converting to numeric
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True).astype(float)
print("Cleaned Phone Column:\n", df)
print("Data types:\n", df.dtypes)
```

```python
# Assuming 'people.csv' is your dataset file path
people_df = pd.read_csv('people.csv', skiprows=50)

# a) Read specific columns
filtered_df = people_df[['Last Name', 'Gender', 'Email', 'Phone', 'Salary']]
print("Filtered DataFrame:\n", filtered_df.head(10))

# b) Extract 'Salary' column as Series and display its last 5 values
salary_series = filtered_df['Salary']
print("Last 5 salaries:\n", salary_series.tail(5))

# c) Filter and select rows where 'Last Name' contains 'Duke', 'Gender' contains 'Female', and 'Salary' < 85000
filtered_rows = filtered_df[(filtered_df['Last Name'].str.contains('Duke')) &
                            (filtered_df['Gender'].str.contains('Female')) &
                            (filtered_df['Salary'] < 85000)]
print("Filtered Rows:\n", filtered_rows)
```

```python
import pandas as pd

try:
    # Load the dataset
    people_df = pd.read_csv('people.csv')

    # a) Try deleting the specified columns, ignore if they don't exist
    columns_to_drop = ['Email', 'Phone', 'Date of birth']
    people_df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

    # b) Drop rows with any missing values
    people_df.dropna(inplace=True)

    # Print the first few rows of the final DataFrame as a check
    print("Final DataFrame (first 5 rows):\n", people_df.head())

except FileNotFoundError:
    print("The file 'people.csv' was not found. Please check the file path.")
except KeyError as e:
    print(f"Column not found: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

```

```python
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='red', marker='o', label='Data points')
plt.axhline(0.5, linestyle='--', color='blue', label='y = 0.5')
plt.axvline(0.5, linestyle=':', color='green', label='x = 0.5')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')
plt.legend()
plt.show()
```

```python
dates = pd.date_range('2023-01-01', periods=100)
temperature = np.random.uniform(10, 40, 100)
humidity = np.random.uniform(20, 80, 100)

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.randn(1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay PDF
xmin, xmax = plt.xlim()  # Get the range for x-axis
x = np.linspace(xmin, xmax, 100)  # Create x values
p = norm.pdf(x, np.mean(data), np.std(data))  # Calculate the PDF

plt.plot(x, p, 'k', linewidth=2)  # Plot the PDF on the histogram

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.show()
```
